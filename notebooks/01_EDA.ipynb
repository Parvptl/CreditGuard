```python
# Import the pandas library
import pandas as pd

# Define the path to your data file
file_path = 'D:/PROJECTS/CreditGuard/data/heloc_dataset_v1.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# --- Initial Inspection ---

# 1. Display the first 5 rows to see the data's structure
print("First 5 rows of the data:")
display(df.head())

# 2. Get a summary of the data types and non-null values
print("\nData types and missing values:")
df.info()

# 3. Get basic statistical details for numerical columns
print("\nStatistical summary:")
display(df.describe())
```

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots for a nicer look
sns.set_style('whitegrid')
```

```python
# --- Analyze the Target Variable ---
plt.figure(figsize=(6, 5))
sns.countplot(x='RiskPerformance', data=df)
plt.title('Distribution of Loan Outcomes (Good vs. Bad/Default)')
plt.xlabel('Loan Status')
plt.ylabel('Number of Applicants')
plt.show()

# Calculate the exact percentage
percentage = df['RiskPerformance'].value_counts(normalize=True) * 100
print(f"Percentage of each class:\n{percentage}")
```

```python
# --- Analyze Numerical Features ---
# Get a list of all columns with numerical data types
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Create histograms for each numerical column
df[numerical_cols].hist(figsize=(15, 12), bins=30, layout=(5, 5))
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.tight_layout()
plt.show()
```

### Key EDA Findings (Initial)

Target Variable: The dataset is well-balanced with a nearly 50/50 split between 'Good' and 'Bad' outcomes. This is excellent, as it means accuracy will be a meaningful metric for our model, and we likely won't need complex resampling techniques like SMOTE.

Numerical Features: Many features, like MSinceOldestTrade, are right-skewed. This suggests that a log transformation may be useful later during preprocessing to improve the performance of linear models.

```python
# --- Bivariate Analysis: Numerical Features ---

# Select only the first 10 numerical columns for a cleaner plot
numerical_subset = numerical_cols[:10]

# Create a grid of boxplots
plt.figure(figsize=(18, 15))
for i, col in enumerate(numerical_subset):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(x='RiskPerformance', y=col, data=df)
    plt.title(f'{col} vs. Loan Outcome')
    plt.xlabel('Loan Status')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# --- Preprocessing Step 1: Handle Special Values ---
import numpy as np

# Replace special codes (-9, -8, -7) with NaN (Not a Number)
df.replace([-9, -8, -7], np.nan, inplace=True)

print("Number of missing values after replacing special codes:")
print(df.isnull().sum())
```

```python
# --- Preprocessing Step 2: Impute Missing Values ---

# Fill NaN values with the median of each column
for col in df.columns:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

print("\nChecking for any remaining missing values:")
print(df.isnull().sum().any()) # Should print False
```

```python
print(df.columns)
```

```python
# --- Preprocessing Step 3: Encode Categorical Features (Robust Method) ---

# First, handle the target variable separately
# This line remains the same
df['RiskPerformance'] = df['RiskPerformance'].apply(lambda x: 1 if x == 'Bad' else 0)

# Now, find all remaining columns that are text-based (categorical)
categorical_cols = df.select_dtypes(include=['object']).columns

# One-hot encode all of them at once
df = pd.get_dummies(df, columns=categorical_cols, prefix=categorical_cols, drop_first=True)

print("\nFirst 5 rows after robust encoding:")
display(df.head())

print("\nColumns in the dataframe after encoding:")
print(df.columns)
```
